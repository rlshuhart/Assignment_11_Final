```python
from yahoo_finance import Share
Share('AMZN').get_price()
```

```python
import pandas as pd
tickers = pd.read_csv("./data/sp100tickers.csv")
tickers.head()
```

```python
prices = []

for tick in tickers['Ticker']:
    print("Retrieving: "+tick + "...")
    p = Share(tick).get_price()
    prices.append(float(p))

prices
```

```python
pd.concat([tickers,pd.DataFrame({"Price":prices})], axis=1).to_csv("./data/sp100tickers_wprices.csv")
```

```python
df = (pd.concat([tickers,pd.DataFrame({"Price":prices})], axis=1)
      .rename(columns={'Name':'name', 'Weight':'size', 'Price':'price'})
     )
df.head()
```

```python
df[['Name','Weight','Price']].to_dict(orient='records')
```

```python
for sector in df['Sector'].unique():
    print(sector)
    print(df[df['Sector']==sector][['name','size','price']].to_dict(orient='records'))
```
